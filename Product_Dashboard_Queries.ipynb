## Product Performance Dashboard Queries
All queries below power the e-commerce product dashboard. Run each cell independently.

### Query 1 — Top 10 Products by Revenue

```python
%sql
SELECT product_name, category, brand,
       ROUND(total_revenue, 2)       AS total_revenue,
       total_units_sold,
       ROUND(avg_return_rate_pct, 2) AS return_rate_pct,
       product_label,
       rank_by_revenue
FROM databricks_cat.gold.ProductPerformance
ORDER BY total_revenue DESC
LIMIT 10
```

### Query 2 — Slow Movers (Review Inventory)

```python
%sql
SELECT product_name, category, brand,
       total_units_sold,
       ROUND(total_revenue, 2)       AS total_revenue,
       ROUND(avg_return_rate_pct, 2) AS return_rate_pct,
       active_selling_days,
       ROUND(revenue_per_day, 2)     AS revenue_per_day
FROM databricks_cat.gold.ProductPerformance
WHERE product_label = 'SLOW_MOVER'
ORDER BY total_units_sold ASC
```

### Query 3 — Revenue by Category

```python
%sql
SELECT category,
       ROUND(total_category_revenue, 2)  AS total_revenue,
       total_category_units,
       distinct_products,
       ROUND(avg_category_return_rate, 2) AS avg_return_rate_pct
FROM databricks_cat.gold.CategoryRevenue
ORDER BY total_revenue DESC
```

### Query 4 — Weekly Demand Trend (Last 12 Weeks)

```python
%sql
SELECT year, week, product_name, category,
       weekly_units_sold,
       ROUND(weekly_revenue, 2)          AS weekly_revenue,
       ROUND(weekly_return_rate_pct, 2)  AS return_rate_pct
FROM databricks_cat.gold.WeeklyDemand
ORDER BY year DESC, week DESC
LIMIT 500
```

### Query 5 — 30-Day Demand Forecast per Product

```python
%sql
SELECT product_name, category,
       forecast_date,
       forecasted_units,
       forecasted_revenue,
       forecast_lower_bound,
       forecast_upper_bound
FROM databricks_cat.gold.DemandForecast_30D
ORDER BY product_name, forecast_date ASC
LIMIT 300
```

### Query 6 — Demand Trend Signals (Rising / Falling / Stable)

```python
%sql
SELECT product_name, category, brand,
       order_date,
       units_sold,
       ROUND(ma_7day_units, 2)  AS ma_7day,
       ROUND(ma_30day_units, 2) AS ma_30day,
       demand_trend
FROM databricks_cat.gold.DemandForecast_MA
ORDER BY product_name, order_date DESC
LIMIT 500
```

### Query 7 — Return Rate Analysis by Product

```python
%sql
SELECT product_name, category, brand,
       total_units_sold,
       ROUND(avg_return_rate_pct, 2) AS avg_return_rate_pct,
       CASE
           WHEN avg_return_rate_pct >= 20 THEN 'HIGH RETURN RISK'
           WHEN avg_return_rate_pct BETWEEN 10 AND 20 THEN 'MODERATE'
           ELSE 'LOW'
       END AS return_risk_label
FROM databricks_cat.gold.ProductPerformance
ORDER BY avg_return_rate_pct DESC
```
